### Import stuff

```python
import nudobe
from nudobe import EFT
from EFT import LEFT
```

### Get Half-Life Ratios

```python
#Set WCs
WC = {"VR(6)": 1e-7}
#Initiate model
model = LEFT(WC, method = "SM")
```

```python
#Define reference isotope
reference_isotope = "76Ge"
```

```python
#Calculate half-life ratios R/R_mbb
ratios = model.ratios(reference_isotope)
ratios
```

```python
#Calculate ratios R/R_mbb with varied LECs
ratios_vary = model.ratios(reference_isotope, vary_LECs = True)
ratios_vary
```

```python
#Get central ratio values of the LEC variation
ratios_vary.median()
```

```python
#Plot Ratios
model.plot_ratios("76Ge", vary_LECs = True, n_points = 1000, alpha = 0.1,
                  show_central = True, savefig = True, file = "CVR6_ratios.png",
                  dpi = 300)
```
