### Study Electron Kinematics

```python
#Set up a LEFT model
WC = {"m_bb" : 0.1e-9,
      "VR(6)" : 3e-7
     }

model = LEFT(WC, method = "SM",
             name  = r"$m_{\beta\beta} + C_\mathrm{VR}^{(6)}$")
```

```python
e = 1e-5 #avoid poles

#Define normalized energy array
Ebar = np.linspace(0+e, 1-e, 10)
```

```python
#Get the spectrum
model.spectrum(Ebar)
```

```python
#GEt angular correlation
model.angular_corr(Ebar)
```

```python
#Make a plot of the spectrum comparing to the standard mechanism
model.plot_spec(linewidth = 4, normalize_x = True, savefig = True,
file = "spec.png")
```

```python
#Make a plot of the angular correlation comparing to the standard mechanism
model.plot_corr(linewidth = 4, normalize_x = False, 
                savefig = True,file = "corr.png")
```
