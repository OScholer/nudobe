### Make plots

```python
fig = LEFT({}, method = "SM").plot_t_half(savefig = True, file = "t_half_mass.png")
```

```python
#Define a model
WC = {"m_bb" : 0.1e-9,
      "VR(6)" : 1e-7
     }
model = LEFT(WC, method = "SM")
```

```python
isotope = "76Ge"
experiments = {"GERDA" : {"limit"     : 1.8e+26, 
                          "color"     : "y", 
                          "linestyle" : "-", 
                          "linewidth" : 1, 
                          "alpha"     : 0.5, 
                          "fill"      : True, 
                          "label"     : "GERDA"
                         }
              }

fig = model.plot_t_half(isotope = isotope, experiments = experiments,
                        show_mbb = True, y_min = 1e+24, y_max = 1e+31,
                        savefig = True, file = "t_half_model.png")
```

```python
#variation of the Unknown LECs in the standard mass mechanism
WC = {"m_bb" : 0.1e-9}
method = "SM"
model = LEFT(WC, method = method)
```

```python
model.plot_t_half_scatter(vary_LECs = True, show_mbb = True,
                          savefig = True, file = "mbb_scatter.png", dpi = 300)
```

```python
#We can also vary different WCs on the x-axis
model.plot_t_half(xaxis = "VR(6)", x_min = 1e-8, x_max = 1e-5,
                  savefig = True, file = "t_half_CVR6.png", dpi = 300)
```

```python
model.plot_t_half_scatter(xaxis = "VR(6)", x_min = 1e-8, x_max = 1e-5, 
                          savefig = True, file = "t_half_scatter_CVR6_vary_phase.png", 
                          dpi = 300)
```

```python
#Vary the LECs additionaly to the phase
model.plot_t_half_scatter(xaxis = "VR(6)", x_min = 1e-8, x_max = 1e-5, vary_LECs = True,
                          savefig = True, file = "t_half_scatter_CVR6_vary_all.png", 
                          dpi = 300)
```

```python
#vary only the LECs and keep the phase
model.plot_t_half_scatter(xaxis = "VR(6)", x_min = 1e-8, x_max = 1e-5, 
                          vary_phases = False, vary_LECs = True,
                          savefig = True, file = "t_half_scatter_CVR6_vary_LEC.png", 
                          dpi = 300)
```
