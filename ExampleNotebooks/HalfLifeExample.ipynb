### Calculate Half-Lives for a specific model

```python
#Define Wilson Coefficients for the standard mass mechanism with m_bb =100meV
WC = {"m_bb" : 0.1e-9}
#NME method
method = "IBM2"
#initiate model
mass_mechanism = LEFT(WC, method = method)
```

```python
#half-life for 76Ge
mass_mechanism.t_half("76Ge")
```

```python
#half-lives in all isotopes
mass_mechanism.half_lives()
```

```python
#half-lives in all isotopes with varied LECs
mass_mechanism.half_lives(vary_LECs = True, n_points = 100)
```
